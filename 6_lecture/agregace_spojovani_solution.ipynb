```python
# 1
stud1 = pd.read_csv('studenti1.csv')
stud2 = pd.read_csv('studenti2.csv')
studenti = pd.concat([stud1, stud2], ignore_index=True)

studenti
```

```python
# 2 
# a) chybi rocnik -> nestuduje
print(len(studenti[studenti['ročník'].isnull()]))

# b) chybi skupina -> dalkovy student
print(len(studenti[studenti.kruh.isnull()]))
```

```python
# 3 Vyčistěte data od studentů, kteří nestudují nebo studují jen dálkově. Nadále budeme pracovat pouze s prezenčními studenty.

studenti = studenti.dropna(subset=['ročník', 'kruh'])
studenti
```

```python
# 4 Zjistěte, kolik prezenčních studentů je v každém z oborů.

studenti.groupby('obor')['kruh'].count()
```

```python
# 5 Zjistěte průměrný prospěch studentů v každém oboru.

studenti.groupby('obor')['prospěch'].mean()
```

```python
# 6 Načtěte datový set s křestními jmény. Proveďte join s tabulkou studentů tak, abychom věděli pohlaví jednotlivých studentů.
jmena = pd.read_csv('jmena.csv')
spojene = pd.merge(studenti, jmena, how='left', on=['jméno'])
spojene
```

```python
# 7 Zjistěte, zda na naší fakultě studují IT spíše ženy nebo spíše muži.
```

```python
spojene.groupby('pohlaví')['jméno'].count()
```
